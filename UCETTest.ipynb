```python
def print_stuff(const, units): #Prints outputs for step 1 and 2
    from uncertainties import ufloat

    data = {} #Creates a dictionary of the two dictionaries
    num = {}
    for name in const:
        try:
            number = round(float(const[name].n),2)
            unc = round(float(const[name].s),2)
            num[name] = ufloat(number, unc)
            try:
                num[name] = (round(float(const[name])),2)
            except:
                num[name] = (const[name])
        except:
            num[name] = (const[name])
    data['values'] = num
    data['units'] = units
    df = pd.DataFrame(data) #Convert to pandas data frame
    # Round all numbers to two decimal places
    #df = df.round(2)

    # Round all numbers to two decimal places
    df = df.round(2)

    # Convert DataFrame values to strings with desired formatting
    df_formatted = df.applymap(lambda x: '{:.2f}'.format(x) if isinstance(x, (int, float)) else x)

    # Apply style and display the table
    df_styled = df_formatted.style.set_properties(**{'border': '1px solid black'})
    display(HTML(df_styled.to_html()))
```

```python
def print_L2(trial, units, average, data_values, uncertainty, N, stadev, interval, high_tier, low_tier, COV): #Print outputs for step 3
    trial['units'] = units #Adds to trial dictionary. Trial dictionary is already a dictionary of dictionaries for each test
    trial['average'] = average
    trial['uncertainty'] = uncertainty
    trial['N'] = N
    trial['standered dev'] = stadev
    trial['interval'] = interval
    trial['high tier'] = high_tier
    trial['low tier'] = low_tier
    trial['COV'] = COV
    
    #Creating list of names to loop through
    names = list_testname
    names.append('units')
    names.append('average')
    names.append('uncertainty')
    names.append('N')
    names.append('standered dev')
    names.append('interval')
    names.append('high tier')
    names.append('low tier')
    names.append('COV')
    
    #Establishing list of important values
    important = ['region',
                'stove_type',
                'stove_height',
                'eff_w_char',
                'burn_rate',
                'firepower']
    
    if average.get('stove_height') == None:
        important[2] = 'combustion_chamber'

        
    display(Markdown('# Cut Table'))
    cut = pd.DataFrame.from_dict(trial, orient='index')[important]
    trans_cut = cut.transpose()
    
    cut_styled = trans_cut.style.set_properties(**{'border': '1px solid black'})
    display(HTML(cut_styled.to_html()))
    
    display(Markdown('# Compared Values')) #Display text as title
    full = pd.DataFrame(trial) #Convert to pandas data frame
    # Apply style and display the table
    full_styled = full.style.set_properties(**{'border': '1px solid black'})
    display(HTML(full_styled.to_html()))
    
```

```python
def load_calcs(): #For running the make input function
    for t in range(len(list_input)): #Run funtion for each entered test
        with output1: #Put in output
            print('')
            print('Looking at test: ' + list_directory[t])
        inputpath = list_input[t] #Look to excel file
        outputpath = os.path.join(list_directory[t], list_testname[t] + '_EnergyInputs.csv') #Output csv
        #Function converts excel data entry sheet into csv sheet, returns a dictionary of values and units
        units, const = LEMS_MakeInputFile_EnergyCalcs(inputpath, outputpath, logpath)
        display(Markdown('# Entered Values')) #Display text as title
        display(Markdown('## For test: ' + list_testname[t])) #Display text as subtitle. Indicate which test it is for
        print('')
        print_stuff(const, units) #Send returned dictionaries to be printed as table
    return units, const

def on_load_clicked(Load): #Function for what to do when button is clicked
    # Clear previous output
    output1.clear_output(wait=True)
    
    with output1:
        load_calcs() #Assign new output from make input file
        
        
    # Update button style to green to track steps done
    Load.style.button_color = 'limegreen'
    Load.description = "Step 1: Load Data"

if test_type == 's' or test_type == 'S' or test_type == 'm' or test_type == 'M':
    output1 = widgets.Output() #Make output
    Load = widgets.Button(description="Step 1: Load Data") #Make button
    Load.on_click(on_load_clicked) #When clicked, run function
    display(Load)
    display(output1)
```
